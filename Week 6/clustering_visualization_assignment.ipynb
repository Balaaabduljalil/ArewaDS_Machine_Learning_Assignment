```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nigerian-songs.csv')
df.head()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
p = figure(title='Bokeh Scatterplot of Dataset')
p.scatter('popularity', 'danceability', source=df, color='artist_top_genre')
show(p)
```

```python
sns.FacetGrid(df,hue='artist_top_genre') \
    .map(plt.scatter,'popularity','danceability') \
    .add_legend()
```

```python
colors = ['red', 'blue']
for label, color in zip(df['artist_top_genre'].unique(), colors):
    subset = df[df['artist_top_genre'] == label]
    plt.scatter(subset['popularity'], subset['danceability'], color=color, label=label)
plt.xlabel('popularity')
plt.ylabel('danceability')
plt.title('scatter plot of dataset.')
plt.legend()
plt.show()
```
