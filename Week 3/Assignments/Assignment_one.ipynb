```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
birds.plot(kind='scatter',x='MaxLength',y='MinLength',figsize=(12,8))
plt.title('Max Length vs Min Length')
plt.ylabel('Min Length')
plt.xlabel('Max Length')
plt.show()
```

```python
birds.plot(x="Family",kind="bar",stacked=True,title="Birds Of Minnesoto")
```
