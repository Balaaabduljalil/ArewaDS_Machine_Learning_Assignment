```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('honey.csv')
honey.head()
```

```python
sns.relplot(data=honey,
            x='year',y='totalprod',
            col='state',
            col_wrap=3,
            kind='line')
```

```python
sns.relplot(data=honey,
            x='yieldpercol',y='prodvalue',
            col='year',
            col_wrap=3,
            kind='line')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['totalprod'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('total production vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['stocks'], ax=ax2, color="r", 
                         label ='stocks', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend();
```
