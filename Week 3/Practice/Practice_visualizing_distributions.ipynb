```python
import pandas as pd 
import matplotlib.pyplot as plt
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
birds.plot(kind='scatter',x='MinLength',y='MaxLength',figsize=(12,8))
plt.title('Max Length per Order')
plt.ylabel('Order')
plt.xlabel('Max Length')

plt.show()
```

```python
birds['MaxBodyMass'].plot(kind='hist',bins=10,figsize=(12,12))
plt.show()
```

```python
birds['MaxBodyMass'].plot(kind='hist',bins=30,figsize=(12,12))
plt.show()
```

```python
df=birds[(birds['MaxBodyMass']>1)&(birds['MaxBodyMass']<60)]
df['MaxBodyMass'].plot(kind='hist',bins=40,figsize=(12,12))
plt.show()
```
